```python
import numpy as np
import pandas as pd
gelirdagılımı = data[["native.country", "capital.gain", "capital.loss"]]
a=gelirdagılımı.groupby("native.country").aggregate(np.sum).reset_index()
a["Net Gelir Sütunu"]= a["capital.gain"] - a["capital.loss"]
a.sort_values(by="Net Gelir Sütunu",ascending=False,inplace=True)
silinecek = a[a["Net Gelir Sütunu"] == 0]
a = a.drop(silinecek.index)
a["Ülke Hal Tablosu"] = None
def Haldurumu(row):
    if row["Net Gelir Sütunu"] < 0:
       return "Fakir"
    elif row["Net Gelir Sütunu"] > 100000:
       return "Zengin"
    else:
       return "Orta Halli"
    
a["Ülke Hal Tablosu"] = a.apply(Haldurumu, axis=1)
a.replace('Hong','Hong-Kong',inplace=True)
a
```

```python
#logistic regresyon
from sklearn import preprocessing
X = data.drop(['income','native.country'], axis=1)
y = data['income']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)


categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex']
for feature in categorical:
        le = preprocessing.LabelEncoder()
        X_train[feature] = le.fit_transform(X_train[feature])
        X_test[feature] = le.transform(X_test[feature])


X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X.columns)

X_test = pd.DataFrame(scaler.transform(X_test), columns = X.columns)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print('Logistic Regression doğruluk puanı: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
